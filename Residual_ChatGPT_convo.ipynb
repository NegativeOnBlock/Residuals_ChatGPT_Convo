```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for X and Y
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
Y = 4 + 3 * X + np.random.randn(100, 1)

# Plot the dataset
plt.scatter(X, Y, color='blue', label='Data Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dataset with Linear Relationship')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate random data for X and Y
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
Y = 4 + 3 * X + np.random.randn(100, 1)

# Fit a Linear Regression Model
model = LinearRegression()
model.fit(X, Y)

# Make Predictions
Y_pred = model.predict(X)

# Plot the dataset and the regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dataset with Linear Relationship and Regression Line')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for X and Y
np.random.seed(42)
X = 1 * np.random.rand(10, 1)
Y = 2 - 10 * X + np.random.randn(10, 1)

# Plot the dataset
plt.scatter(X, Y, color='blue', label='Data Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dataset with Linear Relationship')
plt.legend()
plt.show()
```

```python
# Generate random data with error term
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
epsilon = np.random.randn(100, 1)
Y_with_error = 4 + 3 * X + epsilon

# Plot the data
plt.scatter(X, Y_with_error, color='blue', label='Data with Error Term')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression with Error Term')
plt.legend()
plt.show()
```

```python
# Generate data without error term
Y_without_error = 4 + 3 * X

# Plot the data
plt.scatter(X, Y_without_error, color='red', label='Data without Error Term')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression without Error Term')
plt.legend()
plt.show()
```
